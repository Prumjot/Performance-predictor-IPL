```python
def cleaning(path):
    # reading with /t right now
    df = pd.read_csv(path,sep="/t")

    #extracting date
    match_date = df.iloc[4]
    date = match_date[0]
    date  = date.split('date,')[1]

    #extracting season
    match_season = df.iloc[3]
    season = match_season[0]
    season  = season.split('season,')[1]

    #cleaning data
    innings = df[19:]
    match_df= innings['version,1.3.0'].str.split(',', expand = True).add_prefix('sec')
    match_df.rename(columns = {'sec1': 'balls', 'sec2':'over_ball', 
                            'sec3':'batting_team', 'sec4':'batsman_striker', 
                            'sec5':'batsman_non_striker', 'sec6':'bowler', 
                            'sec7':'runs_scored', 'sec8':'extras?', 
                            'sec9':'dissmisal_type','sec10':'batsman_out'}, 
                            inplace = True)
    match_df.drop(['balls', 'sec0'] , axis = 1, inplace = True)

    #add season and date
    match_df['date'] = date
    match_df['season']= season

    return match_df
```

```python
def concating_dataframes(path):
    '''input: path to the folder with csv files
        output: concatenated all the files in one'''
    path = '../../capstone_project/ipl_csv/Data/trial/'
    allFiles = glob.glob(path + "/*.csv")
    frame = pd.DataFrame()
    list_ = []
    for file_ in allFiles:
        df = pd.read_csv(file_,index_col=None, header=0)
        list_.append(df)
    frame = pd.concat(list_)
    return frame
```

```python
def runs_per_season_trail(df):
    #df_1 = extracting_total_outs_per_batsman(df)
    df['ball']=1
    df= (df.groupby(['batsman_striker','match_id']).sum()).reset_index()
    df['innings']=1
    df_sum = df.pivot_table(index='batsman_striker',values=['runs_scored', 'innings', 'ball'],aggfunc=sum)
    df_sum= df_sum.reset_index()
    df_max = df.pivot_table(index='batsman_striker',values=['runs_scored'],aggfunc=max)
    df_max = df_max.reset_index()
    df_sum = df_sum.merge(a, on ='batsman_striker').merge(df_max, on='batsman_striker')
    return df_sum
```
